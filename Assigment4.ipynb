```python
class Node: 
  def __init__ (self, data):
    self.data = data;
    self.leftChild = None;
    self.rigthChild = None;

```

```python
class BinarySearchTree:
    def __init__(self):
        self.root = None;
        
    def returnRoot(self):
        return self.root
    def insert (self, data):
        if not self.root: # in case if there are no root, data will be rott
          self.root = Node(data);
        else:
          self.insertNode (data, self.root);
  
    def insertNode (self, data, node):
        if data < node.data:
          if node.leftChild: # if there are left chelud...
            self.insertNode(data, node.leftChild);
          else:
            node.leftChild = Node(data)
        if data > node.data:
          if node.rigthChild:
            self.insertNode(data, node.rigthChild);
          else:
            node.rigthChild = Node(data);
                
        
        #Next is maximum and minimum value
        #If we want to get max, we have to go to the right untill the end.
        #In case with min, we go to the left untill the end
        
        
    def getMinValue(self):
        if self.root:
            return self.getMin(self.root)

    def getMin(self, node):
        if node.leftChild:  # if there are left child...
            return self.getMin(node.leftChild) # recursively calling funtion
                # untill it reaches the most left leaf
    
        return node.data # return value of leaf
        
        
        
    def getMaxValue(self):
            if self.root:
                return self.getMax(self.root)
            
    def getMax(self, node):
            if node.rigthChild:
                return self.getMax(node.rigthChild) # recursively calling funtion
                # untill it reaches the most right leaf
            
            return node.data # return value of the most left leaf
        
     

    def traverse(self):
            if self.root:
                return self.traverseInOrder(self.root)
        
    def traverseInOrder(self, node):
            if node.leftChild:
                self.traverseInOrder(node.leftChild)
                
            print(node.data)
            
            if node.rigthChild:
                self.traverseInOrder(node.rigthChild)
                
    # note removing
    
    def remove(self, data):
        if self.root:
            self.root = self.removeNode(data, self.root)
            
    def getPredecessor(self, node): # function to find the largest elemnt in subtree
        if node.rigthChild:
            return self.getPredecessor(node.rigthChild)
        return node
    
    
    def removeNode(self, data, node):
        if not node: # in case if node is NONE
            return node
        
        if data < node.data:
            node.leftChild = self.removeNode(data, node.leftChild)
        elif data > node.data:
            node.rigthChild = self.removeNode(data, node.rigthChild) #     
            #x      
        else: #    
            
            
            if not node.leftChild and not node.rigthChild: # deleting node without child
                print('Removing leaf node')
                del node
                return None
            if not node.leftChild: # removing node where are no left child
                print('Removing node with single right child')
                tempNode = node.rigthChild # creating temp variable,   
                del node
                return tempNode
            
            elif not node.rigthChild: # removing node where are no right child
                print('Removing node wuth single left child')
                tempNode = node.leftChild
                del node
                return tempNode
            
            #        ,   ,      
            
            print('Deleting node with two children')
            tempNode = self.getPredecessor(node.leftChild) # taking the max value of left subtree
            node.data = tempNode.data #        
            node.leftChild = self.removeNode(tempNode.data, node.leftChild) 
            
        return node
    
    def findParent(self, node, value, parent=None):
        if node is None:
            return
        
        if node.data == value:
            print(parent)
        self.findParent(node.leftChild, value, node.data)
        self.findParent(node.rigthChild, value, node.data)
                    
        
    
    
```

```python
bst = BinarySearchTree()
bst.insert(10)
bst.insert(5)
bst.insert(15)
bst.insert(6)
print(bst.getMinValue())
print(bst.getMaxValue())
bst.traverse()
```
